```python
class Employee:
    def __init__(self, name, position, salary):
        self.name = name
        self.position = position
        self.salary = salary

    def display_info(self):
        return f"{self.name} is a {self.position} earning ${self.salary} annually."
```

```python
employee1 = Employee("John Smith", "Manager", 75000)
employee2 = Employee("Jane Doe", "Developer", 60000)
```

```python
print(employee1.display_info())
```

```python
print(employee2.display_info())
```

```python
employee1.salary = 100000
print(employee1.display_info())
```
