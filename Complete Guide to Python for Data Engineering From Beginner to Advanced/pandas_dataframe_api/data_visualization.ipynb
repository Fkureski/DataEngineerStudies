```python
import pandas as pd

import matplotlib as plt
```

```python
# Creating new data
data = {'Date' : ['2025-01-25', '2025-05-23', '2025-10-27'],
        'Price' : [100, 200, 300]}

df = pd.DataFrame(data)

```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.plot(x = 'Date', y = 'Price', kind = 'line')
```

```python
data = {'Product' : ['A', 'B', 'C'],
        'Sales' : [500, 750, 300]}

df = pd.DataFrame(data)

df.plot(x='Product', y='Sales', kind='bar')
```

```python
data = {'Hours_Studied' : ['2', '4', '5', '7', '8', '10'],
        'Exam_Score' : [60, 70, 75, 85, 90, 100]}

df = pd.DataFrame(data)

df.plot(x='Hours_Studied', y='Exam_Score', kind='scatter')
```

```python
data = {'Topping' : ['Pepperoni', 'Mushroom', 'Olive', 'Sausage'],
        'Count' : [25, 15, 10, 20]}

df = pd.DataFrame(data)

df.plot(x='Topping', y='Count', kind = 'pie')
```
