```python
#Step-by-step:
#   1. Read data: Load the CSV data into a pandas DataFrame
#
#   2. Data cleaning: Check for and remove rows with missing values in any column
#
#   3. Data manipulation and analysis:
#       - Calculate total sales per product
#       - Indetify the best-selling product
#       - Compute average daily sales
#
#   4. Vizualization:
#       - Plot sales trends over time
#       - Display sales per product in a bar chart
#
#   5. Class and function:
#       - Define a Python class RetailSalesAnalyzer with methods for each of the above tasks
#           a. Loading data
#           b. Calculating statistics
#           c. Creating visualizations
#
#   6. Script execution
#       - Write a script that creates an instance of RetailSalesAnalyzer, class its methods, and
#       displays the results
```

```python
import pandas as pd
import matplotlib.pyplot as plt

class RetailsSalesAnalyzer:
    def __init__(self):
        self.data = pd.read_csv('daily_sales.csv', sep='\t')
        self.data["Date"] = pd.to_datetime(self.data['Date'], format='%d/%m/%Y', errors='coerce')

    def clean_data(self):
        return self.data.dropna(inplace=True)

    def sales(self):
        #total_sales = self.data.groupby('Product')['Sales'].sum().reset_index()
        #print(total_sales)
        return self.data.groupby('Product')['Sales'].sum()

    def best_sellers(self):
        best_selling_product = self.data.groupby('Product')['Quantity'].sum().reset_index()
        #print(best_selling_product.idxmax())
        return best_selling_product

    def average_daily_sales(self):
        avg_daily_sales = self.data['Sales'].mean()
        print(f'Average daily sales: {avg_daily_sales}')   

    def daily_summary_graphic(self):
        self.data.groupby('Date')['Sales'].sum().plot(kind = 'line')
        plt.title('Sale Trend over Time')
        plt.ylabel('Date')
        plt.ylabel('Total Sales')
        plt.show()
        #daily_summary = self.data.groupby('Date')['Sales'].sum().reset_index()
        #daily_summary = daily_summary.sort_values('Date')
        #daily_summary.plot(x = 'Date', y = 'Sales', kind = 'line')

    def sales_per_product_graphic(self):
        self.sales().plot(kind = 'bar')
        plt.title('Sales per Product')
        plt.ylabel('Product')
        plt.ylabel('Sales')
        plt.show()
        #sales_per_product = self.data.groupby('Product')['Sales'].sum().reset_index()
        #sales_per_product = sales_per_product.sort_values('Product')
        #sales_per_product.plot(x = 'Product', y = 'Sales', kind = 'bar')
    
```

```python
analyzer = RetailsSalesAnalyzer()
analyzer.clean_data()
print('Total sales per product: \n', analyzer.sales())
print()
print('Best selling product: \n', analyzer.best_sellers())
print()
analyzer.average_daily_sales()
print()
print('Total sales per product: \n', analyzer.sales())
print()
analyzer.daily_summary_graphic()
analyzer.sales_per_product_graphic()
```
