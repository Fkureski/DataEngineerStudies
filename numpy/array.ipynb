```python
import numpy as np # Import NumPy
```

```python
nums = [1,2,3,4,5]

array1 = np.array(nums) # Create an array from list

print(array1) 
```

```python
zero_array = np.zeros(5) # Create an array of five 0s
print(zero_array)
```

```python
one_array = np.ones(3) # Create an array of three 1s
print(one_array)
```

```python
range_array = np.arange(1, 10, 2) # Create an array but with a for loop
print(range_array)
```

```python
random_array = np.random.randint(1, 100, 5) # Create a for loop array but randomly
print(random_array)
```

```python
random_matrix = np.random.rand(3,3) # Create a matrix of size three cross three with random values 
print(random_matrix)
```
